```python
# En esta línea de código se limpian los diferentes valores de Comuna que se tienen
df.MATRICULA_RENOVACION.replace(to_replace =["Matrícula", "Renovación"],  
                            value =["Matricula", "Renovacion"], inplace=True)
# En esta línea se reemplazan las comunas con nombre de municipio por el nombre del municipio
df.TAMANO_ACTIVOS.replace(to_replace =["Microempresa", "Pequeña", "Mediana", "Grande"],  
                            value =["MICRO", "PEQUEÑA", "MEDIANA", "GRANDE"], inplace=True) 
# En esta línea se reemplazan las ciudades diferentes a la jurisdicción de la CC por Cali para facilitar el análisis
df.TAMANO_INGRESOS.replace(to_replace =["Microempresa", "Pequeña", "Mediana", "Grande"],  
                            value =["MICRO", "PEQUEÑA", "MEDIANA", "GRANDE"], inplace=True) 
# En esta línea se reemplazan las ciudades diferentes a la jurisdicción de la CC por Cali para facilitar el análisis
df.DESC_CATEGORIA.replace(to_replace =["P. Natural", "Sociedad", "ESAL"],  
                            value =["PERSONA NATURAL", "PERSONA JURIDICA", "ESAL"], inplace=True)
```

```python
#Se convierte la variable FECHA_MATRICULA, FECHA_RENOVACION y FECHA_CANCELACION en datetime para facilitar manupulación de los datos
df.FECHA_MATRICULA = pd.to_datetime(df.FECHA_MATRICULA, format='%Y%m%d', errors='coerce')
df.FECHA_RENOVACION = pd.to_datetime(df.FECHA_RENOVACION, format='%Y%m%d', errors='coerce')
df.FECHA_CANCELACION = pd.to_datetime(df.FECHA_CANCELACION, format='%Y%m%d', errors='coerce')
```

```python
# Para obtener los duplicados EXCEPTO la primera observación revisando todas las columnas 
duplicados = df[df.duplicated()]
print('El número de registros duplicados con todas las columnas iguales son:')
print (duplicados.MATRICULA.count())
```

```python
# Para obtener los duplicados EXCEPTO la última observación revisando la columna NUMERO_IDENTIFICACION. Es decir, se tomarán todas las matriculas duplicadas excepto la que registró la ultima modificación 
duplicadosMAT = df[df.duplicated(['NUMERO_IDENTIFICACION'], keep='last')]
print("El número de Matriculas duplicadas es:", duplicadosMAT.NUMERO_IDENTIFICACION.count(), sep='\n')
```

```python
# Para obtener los duplicados EXCEPTO la última observación revisando en simultaneo las columnas NUMERO_IDENTIFICACION y MATRICULA_RENOVACION. Es decir, se tomarán todas las matriculas duplicadas tanto con su Nnúmero de matricula como lo que hayan hecho en ese año, es decir, si presentó dos renovaciones o dos matriculas se tomarán en la subase creada. Es necesario acalarar que lo anterior se hace exceptuando la que registró la ultima modificación. Esto puede ocurrir debido a que los tramites de renovación de alguna empresa estuveiron mal y se tradaron más del plazo para entregar sus papeles así que se registra una nueva fecha de renovación sin eliminar la primera.
duplicadosMAT_REN = df[df.duplicated(['NUMERO_IDENTIFICACION', 'MATRICULA_RENOVACION'], keep='last')]
print("El número de filas duplicadas analizando dos columnas es:", duplicadosMAT_REN.NUMERO_IDENTIFICACION.count(), sep='\n')
```

```python
# Antes de partir la base de datos. Se hace un conteo para saber cuantas Matriculas y Renovaciones se tienen en la base sin eliminar duplicados
df.MATRICULA_RENOVACION.value_counts()
```

```python
# Ahora es necesario partir la base de datos en dos para hacer la correspondiente limpieza. Se requieren tratar las empresas Matriculadas y Renovadas diferenciadas por su fecha de registro
df_matricula = df.loc[df['MATRICULA_RENOVACION'].isin(['Matricula'])]
df_renovacion = df.loc[df['MATRICULA_RENOVACION'].isin(['Renovacion'])]
print('El número de registros en la BD para Matriculas es:', df_matricula.MATRICULA_RENOVACION.value_counts(), sep='\n')
print('El número de registros en la BD para Renovaciones:', df_renovacion.MATRICULA_RENOVACION.value_counts(), sep='\n')
```

```python
# Se procede a eliminar los duplicados en las dos bases y la idea es dejar el último movimiento registrado. Esto se debe gracias a que se identificó que algunas empresas registran algun movimiento y por algun motivo este es devuelto, en la mayoría de los casos la corrección sugerida es hecha en el plazo para mantener la fecha inicial del movimiento, sin embargo, cuando la empresa deja pasar este plazo máximo se debe generar un nuevo movimiento lo cual genera duplicados en la información
df2_matricula=df_matricula.sort_values('FECHA_MATRICULA').drop_duplicates('NUMERO_IDENTIFICACION', keep='last')
df2_renovacion=df_renovacion.sort_values('FECHA_RENOVACION').drop_duplicates('NUMERO_IDENTIFICACION', keep='last')
```

```python
# Una vez eliminados lo duplicados se cuenta el número de resgitros 
print('El número de registros en la BD para Matriculas después de eliminar duplicados es:', df2_matricula.MATRICULA_RENOVACION.value_counts(), sep='\n')
print('El número de registros en la BD para Renovaciones después de eliminar duplicados es:', df2_renovacion.MATRICULA_RENOVACION.value_counts(), sep='\n')
```

```python
# Se procede a unir las bases de datos para la posterior limpieza en conjunto
df = pd.concat([df2_matricula, df2_renovacion])
df = df.sort_values('MATRICULA_RENOVACION', ascending=True)
print('El número de registros en la BD unida es:', df.MATRICULA_RENOVACION.value_counts(), sep='\n')

df.sort_values('FECHA_MATRICULA').head()
```

```python
# Para eliminar duplicados se debe tener en cuenta que en NUMERO_IDENTIFICACION nos interesa dejar el registro más reciente que tengamos en la base. Aquí el sort está organizando del más antiguo al más nuevo
df = df.sort_values('FECHA_MATRICULA').drop_duplicates('NUMERO_IDENTIFICACION', keep='last')
print('El número de registros en la BD unida sin duplicados es:', df.MATRICULA_RENOVACION.value_counts(), sep='\n')
```
